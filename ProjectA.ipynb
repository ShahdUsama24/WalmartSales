```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y') # Changed format to '%d/%m/%Y' to match DD/MM/YYYY
# extract year,month,day from date feature for futher analysis
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
df['Day']=df['Date'].dt.day
df['Year-Month']=df['Year'].astype(str) + '-' +df['Month'].astype(str)
```

```python
df['New_Temperature'] = (df['Temperature'] - df['Temperature'].min()) / (df['Temperature'].max() - df['Temperature'].min())
df['New_Temperature'] = np.where(df['Temperature'] >= 0.5, 1, 0)

print(df['New_Temperature'])
```

```python
df['Fuel_Price']=(df['Fuel_Price']-df['Fuel_Price'].min())/(df['Fuel_Price'].max()-df['Fuel_Price'].min())
df['New_Fuel_Price'] = (df['Fuel_Price'] - df['Fuel_Price'].min()) / (df['Fuel_Price'].max() - df['Fuel_Price'].min())
df['New_Fuel_Price'] = np.where(df['Fuel_Price'] >= 0.5, 1, 0)

df['New_Fuel_Price'].value_counts()
```

```python
df['INTUnemployment'] = df['Unemployment'].astype(int)

df['INTUnemployment'].value_counts()
bins=[3,6,9,12,15]
labels=['3-6','6-9','9-12','12-15']
df['Unemployment_rate_bins']=pd.cut(df['INTUnemployment'],bins=bins,labels=labels)
print(df['Unemployment_rate_bins'])
df['Unemployment_rate_bins'].value_counts()
```

```python
df['CPI'] = df['CPI'].astype(int)
print(df['CPI'])
df['CPI'].value_counts()
df['CPI'] = (df['CPI'] - df['CPI'].min()) / (df['CPI'].max() - df['CPI'].min())
print(df['CPI'])
df['CPI_pct_change'] = df['CPI'].pct_change() * 100
df['CPI_pct_change'] = df['CPI_pct_change'].fillna(0)
print(df['CPI_pct_change'])
df['CPI_pct_change'].value_counts()
def custom_bin(value):
    if value == 0:
        return 'Group 1'
    elif value > 1:
        return 'Group 2'
    elif value < 1:
        return 'Group 3'
    else:
        return 'Group 4'

# Applying the custom binning function
df['CPI_bin'] = df['CPI_pct_change'].apply(custom_bin)
df['CPI_bin'].value_counts()
```

```python
bins=np.linspace(min(df['Weekly_Sales']),max(df['Weekly_Sales']),5)
group_names=['Low','Medium','High_Med','High']
df['Weekly_Sales_bins']=pd.cut(df['Weekly_Sales'],bins,labels=group_names,include_lowest=True)
print(df['Weekly_Sales_bins'])
df['Weekly_Sales_bins'].value_counts()
```

```python
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['Season'] = df['Date'].dt.month.apply(month_to_season)
```

```python
Weekly_Sales_Season=df.groupby(['Season'])['Weekly_Sales'].sum().reset_index()
# Use 'Weekly_Sales_Season' for both data and labels to ensure matching lengths
plt.pie(Weekly_Sales_Season['Weekly_Sales'], labels=Weekly_Sales_Season['Season'], autopct='%1.1f%%', startangle=140)
plt.title('Weekly Sales Distribution by Season')
plt.axis('equal')
plt.show()
```
